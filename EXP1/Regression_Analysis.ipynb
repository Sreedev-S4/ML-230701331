```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load Iris dataset from online source
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

# Preview the data
display(df.head())
```

```python
# 1. UNIVARIATE REGRESSION
X_uni = df[['petal_width']]
y_uni = df['petal_length']

model_uni = LinearRegression()
model_uni.fit(X_uni, y_uni)

plt.figure(figsize=(8, 5))
sns.regplot(x='petal_width', y='petal_length', data=df, line_kws={"color": "red"})
plt.title(f'Univariate: Petal Width vs Petal Length (R²: {model_uni.score(X_uni, y_uni):.4f})')
plt.show()
```

```python
# 2. BIVARIATE ANALYSIS
correlation = df['sepal_length'].corr(df['sepal_width'])
print(f"Bivariate Correlation (Sepal Length vs Width): {correlation:.4f}")

plt.figure(figsize=(8, 5))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title('Bivariate Analysis: Sepal Dimensions by Species')
plt.show()
```

```python
# 3. MULTIVARIATE REGRESSION
from mpl_toolkits.mplot3d import Axes3D

X_multi = df[['sepal_length', 'sepal_width']]
y_multi = df['petal_length']

model_multi = LinearRegression()
model_multi.fit(X_multi, y_multi)

# Create 3D Visualization
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the data points
ax.scatter(df['sepal_length'], df['sepal_width'], df['petal_length'], c='blue', alpha=0.6)

# Labels
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_zlabel('Petal Length (cm)')
ax.set_title('Multivariate Regression: 3D Feature Analysis')

# Set the view angle to match your reference
ax.view_init(elev=20, azim=130)

plt.show()

print(f"Multivariate R² Score: {model_multi.score(X_multi, y_multi):.4f}")
```
