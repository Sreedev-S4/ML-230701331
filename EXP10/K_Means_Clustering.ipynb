```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd

# 1. Loading the Iris dataset
# We only use X (features) because K-Means doesn't use labels (y)
iris = load_iris()
X = iris.data 

# 2. Training the K-Means Model
# We set n_clusters=3 because we know there are 3 species of Iris
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)

# 3. Formatted Output for Lab Record
print("Output:")
print(f"\nCluster Assignments: {y_kmeans}")

# The centroids represent the 'average' point of each cluster
print(f"\nCluster Centers (Centroids):\n{kmeans.cluster_centers_}")

# 4. Visualizing the Clusters (Plotting Sepal Length vs Sepal Width)
plt.figure(figsize=(10, 6))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')

# Plotting the Centroids (Yellow Stars)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            s=300, c='yellow', label='Centroids', marker='*')

plt.title('K-Means Clustering (Iris Dataset)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()

print("\nResult: Unsupervised K-Means Clustering Complete!")
```
