```python
import numpy as np

# 1. Define the Step Activation Function
def activation_function(x):
    return 1 if x >= 0 else 0

# 2. Initialize Training Data (AND Gate logic)
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
targets = np.array([0, 0, 0, 1])

# 3. Initialize Weights, Bias, and Learning Rate
weights = np.array([0.0, 0.0])
bias = 0.0
learning_rate = 0.2
epochs = 5

print("--- Training Single Layer Perceptron ---")

# 4. Training Loop (Epoch by Epoch with details)
for epoch in range(1, epochs + 1):
    print(f"\nEpoch {epoch}")
    for i in range(len(inputs)):
        # Calculate Linear Combination: z = w1*x1 + w2*x2 + b
        linear_output = np.dot(inputs[i], weights) + bias
        prediction = activation_function(linear_output)
        
        # Calculate Error: Error = Target - Prediction
        error = targets[i] - prediction
        
        # Update Weights and Bias: w = w + (lr * error * x)
        weights += learning_rate * error * inputs[i]
        bias += learning_rate * error
        
        print(f"Input: {inputs[i]}, Predicted: {prediction}, Error: {error}")

# 5. Final Model Parameters
print("\n" + "="*30)
print(f"Final Weights: {weights}")
print(f"Final Bias: {bias}")
print("="*30)

# 6. Testing the Perceptron
print("\nTesting Perceptron:")
for i in range(len(inputs)):
    test_out = activation_function(np.dot(inputs[i], weights) + bias)
    print(f"Input: {inputs[i]}, Output: {test_out}")
```
