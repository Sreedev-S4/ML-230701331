```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 1. Loading the Iris dataset (Classes 0, 1, 2)
iris = load_iris()
X, y = iris.data, iris.target

# 2. Splitting the dataset
# Using a specific random_state to get results close to your notebook
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

# 3. Training the Decision Tree
classifier = DecisionTreeClassifier(criterion='entropy', random_state=42)
classifier.fit(X_train, y_train)

# 4. Making Predictions
y_pred = classifier.predict(X_test)

# 5. Formatted Output to match your handwritten record
print("Output:")
print(f"\nPredicted Values: {y_pred}")
print(f"\nActual Values: {y_test}")

# Calculating accuracy to high precision as seen in image
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy : {accuracy:.7f}")

print("\nResult:")
```
