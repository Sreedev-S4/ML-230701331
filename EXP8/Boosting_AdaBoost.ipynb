```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report

# 1. Loading the Iris dataset (Classes 0, 1, 2)
iris = load_iris()
X, y = iris.data, iris.target

# 2. Splitting the dataset
# 30% test size to keep it comparable to your Decision Tree results
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

# 3. Training the AdaBoost Classifier
# n_estimators=50: The number of sequential learners
# learning_rate=1.0: Controls the contribution of each weak learner
model = AdaBoostClassifier(n_estimators=50, learning_rate=1.0, random_state=42)
model.fit(X_train, y_train)

# 4. Making Predictions
y_pred = model.predict(X_test)

# 5. Formatted Output (Matching your lab record style)
print("Output:")
print(f"\nPredicted Values: {y_pred}")
print(f"\nActual Values: {y_test}")

# Boosting should yield a very high accuracy (0.9777... or 1.0)
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy : {accuracy:.7f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nResult: Sequential Boosting Training Complete!")
```
